```python
import pandas as pd

data = pd.read_csv("dataset.csv")

try:
    genre_data = pd.read_csv("genre_data.csv")
except:
    genre_data = pd.DataFrame()

try:
    year_data = pd.read_csv("year_data.csv")
except:
    year_data = pd.DataFrame()

try:
    artist_data = pd.read_csv("artist_data.csv")
except:
    artist_data = pd.DataFrame()

print("✅ Data loaded")
```

```python
print("Main Data (first 2 rows):")
display(data.head(2))

print("\nGenre Data (first 2 rows):")
display(genre_data.head(2))

print("\nYear Data (first 2 rows):")
display(year_data.head(2))

print("\nArtist Data (first 2 rows):")
display(artist_data.head(2))
```

```python
print("Info of Main Data:")
print(data.info())

print("\nInfo of Genre Data:")
print(genre_data.info())
```

```python
if "year" in data.columns:
    data["decade"] = data["year"].apply(lambda x: str(x)[:3] + "0s")
    print("✅ Decade column created")
else:
    print("⚠️ Column 'year' not found. Decade column not created.")
```
